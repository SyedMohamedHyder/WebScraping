```python
import json
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://helmboots.com/products.json?limit=250&page=1'
```

```python
response = requests.get(url)
products_json = response.text
```

```python
type(products_json)
```

```python
products = json.loads(products_json)
```

```python
products_list = []
for shoes in products['products']:
    images = {image['product_id']: image['src'] for image in shoes['images']}
    for shoe in shoes['variants']:
        shoe['title'] = shoes['title']
        shoe['handle'] = shoes['handle']
        shoe['product_type'] = shoes['product_type']
        shoe['img'] = images[shoe['product_id']]
        products_list.append(shoe)
```

```python
len(products_list)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(products_list)
```

```python
df.to_csv('shoes.csv')
```
