```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://fastestlaps.com/tracks/monza'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
table_body = soup.find('tbody')
```

```python
table_body
```

```python
car_details = table_body.find_all('tr')
```

```python
cars = []
for car_detail in car_details:
    data = dict(
        rank = car_detail.find('td').text.strip('.'),
        vehicle = car_detail.find_all('a')[0].text.strip(),
        driver = car_detail.find_all('td')[2].text.strip(),
        time = car_detail.find_all('a')[1].text.strip(),
    )
    cars.append(data)
```

```python
len(cars)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cars)
df.to_csv('cars_raking.csv', index=False)
```
