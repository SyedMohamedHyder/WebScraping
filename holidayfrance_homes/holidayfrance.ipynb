```python
import requests
from bs4 import BeautifulSoup
from functools import wraps

def get_page_cottages(page_number):
    
    url = f'https://www.holidayfrancedirect.co.uk/cottages-holidays/index.htm?board=sc&d=France&people=2&prop_type%5B0%5D=cottagegite&page={page_number}'
    
    headers = {
    'authority': 'www.holidayfrancedirect.co.uk',
    'cache-control': 'max-age=0',
    'sec-ch-ua': '^\\^Google',
    'sec-ch-ua-mobile': '?0',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'sec-fetch-site': 'none',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-user': '?1',
    'sec-fetch-dest': 'document',
    'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8'
    }

    response = requests.get(url, headers=headers)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    return soup.find_all('div', {'class': 'col-content'})



def prime_iter(gen_fn):
    @wraps(gen_fn)
    def inner():
        gen = gen_fn()
        next(gen)
        return gen
    return inner


@prime_iter
def get_cottage_data():
    cottage_info = None
    while True:
        cottage = yield cottage_info
        cottage_info = dict(
        name = cottage.find('a').text.split('-')[0].strip(),
        cottage_id = cottage.find('a').text.split('-')[1],
        cottage_image = cottage.find('a')['href'],
        bedrooms = cottage.find('p', {'class': 'property-spec'}).text.split(',')[0],
        sleeps = cottage.find('p', {'class': 'property-spec'}).text.split(',')[1],
        rating = cottage.find('p', {'class': 'property-spec'}).text.split(',')[-1].split()[-1],
        price = cottage.find('div', {'class': 'property-pricing'}).text
    )
        
def get_all_cottages(page_numbers):
    cottage_data_collector = get_cottage_data()
    all_cottages = list()
    for page in range(1, page_numbers+1):
        cottages_page = get_page_cottages(page)
        for cottage in cottages_page:
            all_cottages.append(cottage_data_collector.send(cottage))
    cottage_data_collector.close()
    return all_cottages
```
