```python
from requests_html import AsyncHTMLSession
from bs4 import BeautifulSoup
import asyncio
import nest_asyncio
```

```python
def html_page(page_number):
    if asyncio.get_event_loop().is_running():
        nest_asyncio.apply()
    with AsyncHTMLSession() as session:
        url = f'https://www.beerwulf.com/en-gb/c/all-beers?page={page_number}'
        response = session.get(url)
    response.html.render(sleep=1)
    return response
```

```python
def parse_data(response):
    beer = response.html.xpath('//*[@id="product-items-container"]', first=True)
    return beer
```
